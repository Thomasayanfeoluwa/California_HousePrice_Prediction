```python
import pickle
import numpy as np

# Load the model and scaler
try:
    regresion = pickle.load(open('models/regresion.pkl', 'rb'))
    scaler = pickle.load(open('models/scaler.pkl', 'rb'))
except FileNotFoundError:
    print("Error: Model or scaler file not found.")
    exit(1)

# Test inputs
input_1 = np.array([[8.3252, 41, 6.9841, 1.0238, 322, 2.5556, 37.88, -122.23]])
input_2 = np.array([[2.0, 20, 4.0, 1.2, 500, 3.0, 34.05, -117.68]])

# Scale inputs
input_1_scaled = scaler.transform(input_1)
input_2_scaled = scaler.transform(input_2)

# Predict
print("Prediction for input_1:", regresion.predict(input_1_scaled)[0])
print("Prediction for input_2:", regresion.predict(input_2_scaled)[0])
# Access the coefficients from the Ridge model in the pipeline
print("Model coefficients:", regresion.named_steps['ridge'].coef_)
```
